```python
plt.figure(figsize=(10, 9))
sns.boxplot(data=df, x='type', y='price')
```

```python
sns.catplot(data=df[df['type']=='1к'], x='square', y='price')
```

```python
plt.figure(figsize=(10, 9))
price_more_15 = df[df['price'] >= 15]
sns.boxplot(data=price_more_15, x = 'type', y = 'square')
```

```python
plt.figure(figsize=(10, 9))
price_less_10 = df[df['price'] <= 10]
sns.boxplot(data=price_less_10, x = 'type', y = 'square')
```

```python
plt.figure(figsize=(10, 9))
price_more_7_less_10 = df[(df['price'] <= 10) & (df['price'] >= 7)]
sns.boxplot(data=price_more_7_less_10, x = 'type', y = 'square')
```

```python
price_more_7_less_10['type'].value_counts()
```

```python
plt.figure(figsize=(10, 9))
price_more_3_less_7 = df[(df['price'] <= 7) & (df['price'] >= 3)]
sns.boxplot(data=price_more_3_less_7, x = 'type', y = 'square')
```

```python
price_more_3_less_7['type'].value_counts()
```
